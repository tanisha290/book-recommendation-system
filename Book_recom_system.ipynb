# POPULARITY BASED RECOMMENDATION SYSTEM


on the basis of books which have no. of ratings more than 250

```python
ratings_with_bname=ratings.merge(books,on='ISBN')
```

```python
ratings_with_bname.shape
```

```python
ratings_with_bname.groupby('Book-Title').count()
```

```python
num_rating_df=ratings_with_bname.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
ratings_with_bname['Book-Rating'] = pd.to_numeric(ratings_with_bname['Book-Rating'], errors='coerce')
avg_rating_df=ratings_with_bname.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
# Merge popular_df with the relevant columns from books
# This merge should add 'Book-Author' and 'Image-URL-M' to popular_df
popular_df = popular_df.merge(books[['Book-Title','Book-Author','Image-URL-M']],on='Book-Title')

# Now select the desired columns from the merged DataFrame and drop duplicates
# Ensure all the columns in the list exist in the popular_df after the merge
popular_df = popular_df[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']].drop_duplicates('Book-Title')
```

# Collaborative Filtering Based Recommendation System

based on those books which have more than 50 ratings and on the users who have given atleast 200 ratings

```python
ratings_with_bname
```

```python
x=ratings_with_bname.groupby('User-ID').count()['Book-Rating']>200
literate_users=x[x==True].index
literate_users
```

```python
filtered_rating=ratings_with_bname[ratings_with_bname['User-ID'].isin(literate_users)]
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y==True].index
famous_books
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
similarity_scores
```

```python
np.where(pt.index=='Zoya')[0]
```

```python
def recommend(book_name):
  #index fetch
  index=np.where(pt.index==book_name)[0][0]
  #similarity of that book which has that index
  distances=similarity_scores[index] #an array
  #list(enumerate(similarity_scores[index]) will number the array
  similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]#1st item not taken..as it is the book itself

  data=[]
  for i in similar_items:
    item=[]
    temp_df=books[books['Book-Title']==pt.index[i[0]]];
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
    data.append(item)
  return data
```

```python
recommend('1984')
```
